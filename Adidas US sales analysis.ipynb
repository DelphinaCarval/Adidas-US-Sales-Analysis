# Data cleaning

```python

# Drop the first four rows and first two columns 
df1 = df.drop(index=range(4), columns=df.columns[:1])

# Reset the index
df1 = df1.reset_index(drop=True)

df1.head()
```

```python

# Define a dictionary of old and new column names
new_names = {'Unnamed: 1': 'Retailer',
             'Unnamed: 2': 'Retailer ID',
             'Unnamed: 3': 'Invoice Date',
             'Unnamed: 4': 'Region',
             'Unnamed: 5': 'State',
             'Unnamed: 6': 'City',
             'Unnamed: 7': 'Product',
             'Unnamed: 8': 'Price per Unit',
             'Unnamed: 9': 'Units Sold',
             'Unnamed: 10': 'Total Sales',
             'Unnamed: 11': 'Operating Profit',
             'Unnamed: 12': 'Operating Margin',
             'Unnamed: 13': 'Sales Method'}

# Rename the columns of the dataframe
df1 = df1.rename(columns=new_names)

df1
```

```python

# Convert the invoice date column to datetime format
df1['Invoice Date'] = pd.to_datetime(df1['Invoice Date'])

# Extract the year from the invoice date column and create a new column
df1['Year'] = df1['Invoice Date'].dt.year



```

```python
df1.head()
```

```python
df1.info()
```

Changing data types of columns

```python
cols_to_convert = ['Price per Unit', 'Units Sold', 'Total Sales', 'Operating Profit', 'Operating Margin']

for col in cols_to_convert:
    df1[col] = pd.to_numeric(df1[col])
```

# Plotting

```python
# Group the data by state and calculate total sales for each state
state_sales = df1.groupby('State')['Total Sales'].sum()

# Create a horizontal bar chart of total sales by state
fig, ax = plt.subplots(figsize=(10, 8))
ax.barh(state_sales.index, state_sales.values, height=0.5)
ax.set_xlabel('Total Sales(M$)')
ax.set_ylabel('State')
ax.set_title('Total Sales by State')

# Adjust the x-axis label font size
for label in ax.xaxis.get_ticklabels():
    label.set_fontsize(10)

plt.show()
```

Insight:-
The horizontal bar graph shows the total sales for each state in the US for the Adidas brand. It is evident that New York, California, and Florida are the top-performing states with the highest sales. This information can be useful for Adidas to allocate resources and prioritize marketing efforts to maximize revenue in high-performing states. Additionally, it can help identify opportunities to improve sales in low-performing states and potentially expand the brand's reach.

```python
import calendar

# Calculate monthly sales
monthly_sales = df1.groupby([df1['Invoice Date'].dt.year, df1['Invoice Date'].dt.month])['Total Sales'].sum()
monthly_sales = monthly_sales.groupby(level=1).sum()
plt.plot(monthly_sales.values, color='blue', linewidth=2, marker='o', markersize=6)
plt.grid(True)
month_names = [calendar.month_name[i][:3] for i in range(1, 13)]
plt.xticks(range(12), month_names)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Sales(M$)', fontsize=12)
plt.title('Monthly Sales', fontsize=14)
for i, v in enumerate(monthly_sales.values):
    plt.text(i, v, "${:,.0f}".format(v), color='blue', fontweight='bold')

plt.show()
```

Insight:- 
The line graph shows how well Adidas products are selling each month . By analyzing the chart, We can see that sales are highest in July, August, and December, and lowest in March. This helps Adidas understand why sales are high or low and set goals to improve sales performance. 

Recommendation:- 
For months with highest sales plan marketing campaigns and product releases around these peak months to take advantage of increased customer demand . During these months, promote seasonal items, such as summer sporting apparel in July and August or winter sports shoes in December . whereas for months with lower sales . Consider partnering with other brands or organizations to create limited-time collaborations or events to generate excitement and drive sales . Identify reasons for low sales during these months by analyzing sales data, customer feedback, and market trends . Develop and promote products that are suited to the season or upcoming events during these months.


```python

# Group the data by sales method and calculate total sales for each method
method_sales = df1.groupby('Sales Method')['Total Sales'].sum()

# Create a pie chart of sales method
plt.pie(method_sales.values, labels=method_sales.index, autopct='%1.1f%%')
plt.title('Sales Method')
plt.show()
```

Insight:-
Based on the pie chart, in-store sales for the Adidas brand generate the highest revenue at 39.6%, followed by outlet sales at 32.8%, and online sales at 27.5%. This information can be useful for Adidas to allocate resources and prioritize their sales method, as well as to potentially identify opportunities to improve sales in low-performing methods.

Recommendation :- 
Adidas could think about investing more resources in advertising and refining its in-store sales experience, since it appears to be the preferred mode of sales for customers. To encourage customers to purchase in-store, Adidas may consider giving unique in-store promotions or limited-edition products . Adidas, on the other hand, could consider investing in its online sales platform, as online purchasing is growing increasingly popular among customers and may be a potential development area . To enhance exposure and attract people to its website, Adidas should focus on improving its website's user experience, giving free or cheap delivery, and expanding its online marketing activities.
